# Listing files

```python
import os
files = os.listdir('stories')
```

# Splitting content for one file

```python
import re
def parse_file( fname ):
    f = open(fname)
    text = re.split(r'\n*@highlight\n\n*', f.read())
    f.close()
    return {'news': text[0], 'summaries': [x for x in text[1:] if x]}
```

# Iterating over files and serializing to file

```python
import pickle

master_list = []
for file in files:
    master_list.append(parse_file(os.path.join('stories', file)))
outfile = open('master_list.pkl','wb')
pickle.dump(master_list, outfile)
outfile.close()
```
